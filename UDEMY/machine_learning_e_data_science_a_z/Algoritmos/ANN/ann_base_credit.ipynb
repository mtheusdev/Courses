```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import pickle
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
with open('../../bases/credit.pkl', 'rb') as f:
    X_credit_treinamento, Y_credit_treinamento, X_credit_teste, Y_credit_teste = pickle.load(f)
```

```python
X_credit_treinamento.shape, Y_credit_treinamento.shape
```

```python
X_credit_teste.shape, Y_credit_teste.shape
```

```python
#3 entradas -> 2, 2 -> 1
ann_credit = MLPClassifier(max_iter=1500, verbose=True, tol=0.0000100, solver='adam', activation='relu', hidden_layer_sizes=(2,2))
```

```python
ann_credit.fit(X_credit_treinamento, Y_credit_treinamento)
```

```python
previsoes = ann_credit.predict(X_credit_teste)
```

```python
previsoes
```

```python
Y_credit_teste
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
accuracy_score(Y_credit_teste, previsoes)
```

```python
from yellowbrick.classifier import ConfusionMatrix
cm = ConfusionMatrix(ann_credit)
cm.fit(X_credit_treinamento, Y_credit_treinamento)
cm.score(X_credit_teste, Y_credit_teste)
```

```python
print(classification_report(Y_credit_teste, previsoes))
```
